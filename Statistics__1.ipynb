**Task 1**

Create a population of 2000 values and draw a random sample of 100.
Print the sample mean and population mean.

```python
import random
import numpy as np

# Population
population = list(range(1, 2001))

# Sample
sample = random.sample(population, 100)

# Means
population_mean = np.mean(population)
sample_mean = np.mean(sample)

print("Population Mean:", population_mean)
print("Sample Mean:", sample_mean)
```

# **Data Collection & Organization**

**Frequency Distribution**

Example

```python
blood_groups = ["A","B","AB","O","B","A","O","AB","B","B","A","O"]
```

```python
import pandas as pd

df = pd.DataFrame(blood_groups, columns=["Blood Group"])
freq = df["Blood Group"].value_counts()
rel_freq = df["Blood Group"].value_counts(normalize=True)

freq_table = pd.DataFrame({
    "Frequency": freq,
    "Relative Frequency": rel_freq
})

freq_table
```

# **Tabulation and Classification**
Example

```python
marks = [45, 67, 89, 56, 72, 90, 38, 61, 77]

df = pd.DataFrame(marks, columns=["Marks"])

df["Class"] = pd.cut(df["Marks"],
                     bins=[0,40,60,80,100],
                     labels=["Fail","Pass","Good","Excellent"])

df
```

## **Scatter Plot & Correlation**

```python
import numpy as np

hours = np.array([2,4,6,8,10])
marks = np.array([40,55,65,80,90])

plt.scatter(hours, marks)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.show()

# Correlation
np.corrcoef(hours, marks)
```

## **Task 10**

Create a scatter plot for advertising cost vs sales
Find and interpret the correlation coefficient.

```python
import numpy as np

advertising = np.array([10, 20, 30, 40, 50])
sales = np.array([100, 180, 260, 340, 420])

plt.scatter(advertising, sales)
plt.xlabel("Advertising Cost")
plt.ylabel("Sales")
plt.title("Advertising Cost vs Sales")
plt.show()

# Correlation
correlation = np.corrcoef(advertising, sales)
correlation
```
